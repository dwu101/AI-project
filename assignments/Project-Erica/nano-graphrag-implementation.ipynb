```python
import json
import numpy as np

# Load chunks
docs = []
with open("chunks.jsonl") as f:
    for line in f:
        item = json.loads(line)
        docs.append({"id": item["id"], "text": item["text"]})

# Load embeddings
embs = []
with open("embeddings.jsonl") as f:
    for line in f:
        e = json.loads(line)
        embs.append(e["embedding"])

embs = np.array(embs)

from nano_graphrag import GraphRAG

rag = GraphRAG(docs, embs)
rag.build_graph(k=10)
```
